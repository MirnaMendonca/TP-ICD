# Análise do dataset com as músicas mais populares do TikTok

Esse notebook tem por objetivo documentar o processo de análise exploratória de um dataset com as músicas mais populates da rede social TikTok nos últimos anos. Pretendemos explorar e destrinchar os dados, procurando por relações entre os atributos das músicas e a popularidade delas, tentando estabelecer, principalmente, relações de causalidade (se possível).

```python
import numpy as np
import pandas as pd

import math

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm

plt.rcParams['figure.figsize']  = (8, 5)
plt.rcParams['axes.labelsize']  = 8
plt.rcParams['axes.titlesize']  = 8
plt.rcParams['legend.fontsize'] = 8
plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8
plt.rcParams['lines.linewidth'] = 2

plt.ion()
plt.style.use('seaborn-colorblind')

sns.set_theme(style="whitegrid")
```

```python
# Importando e exibindo as primeiras linhas do dataset

tiktok = pd.read_csv('../data/tiktok.csv')
tiktok.head()
```

```python
print("Número de registros no dataset:", len(tiktok))
```

```python
# Removendo dados NaN, NA, etc

tiktok.dropna(inplace=True)
tiktok.reset_index(drop=True, inplace=True)

print("Número de registros no dataset após remover as linhas com dados faltantes:", len(tiktok))
```

```python
# Lista todas as colunas do dataset

print(tiktok.columns)
```

```python
# Seleciona as colunas que não serão utilizadas e, portanto, removidas

columns = ['Unnamed: 0', 'track_id', 'album_id', 'artist_id', 'playlist_id', 'playlist_name']

tiktok.drop(columns, axis=1, inplace=True)
tiktok.head()
```

```python
# Encontra os registros duplicados e os exibe

duplicates = tiktok[tiktok.duplicated(keep='last')]

print("Número de registros duplicados:", len(duplicates))
duplicates.head()
```

```python
# Remove os registros duplciados

tiktok.drop_duplicates(inplace=True)
tiktok.reset_index(drop=True, inplace=True)

print("Número de registros sem as duplicatas:", len(tiktok))
```

```python
# Renomeia os atributos da coluna "genre" do dataset

print(tiktok['genre'].unique())
tiktok['genre'].replace({'_TIKTOK': 'TIKTOK'}, inplace=True)

print(tiktok['genre'].unique())
```

```python
# Cria a coluna "year" para facilitar agregações

tiktok['release_year'] = [int(item[:4]) for item in tiktok['release_date']]

tiktok.head()
```

```python
min_dur = tiktok['duration_mins'].min()
max_dur = tiktok['duration_mins'].max()
mean_dur = tiktok['duration_mins'].mean()
median_dur = tiktok['duration_mins'].median()

print("Duração mínima (minutos):", round(min_dur, 2))
print("Duração máxima (minutos):", round(max_dur, 2))

tiktok.plot.hist(y='duration_mins', edgecolor='k')
plt.title('Histograma da duração das músicas (em minutos)')
plt.xlabel('Duração')
plt.ylabel('Num. músicas')

plt.axvline(min_dur, color = 'green', linestyle='dashed', label='min')
plt.axvline(mean_dur, color = 'cyan', linestyle='dashed', label='mean')
plt.axvline(median_dur, color = 'red', linestyle='dashed', label='median')
plt.axvline(max_dur, color = 'orange', linestyle='dashed', label='max')

plt.legend(loc=1)
```

```python
min_year = tiktok['release_year'].min()
max_year = tiktok['release_year'].max()
mean_year = tiktok['release_year'].mean()
median_year = tiktok['release_year'].median()

print("Ano de lançamento da música mais antiga:", min_year)
print("Ano de lançamento da música mais recente:", max_year)

tiktok.plot.hist(y='release_year', edgecolor='k')
plt.title('Histograma do ano de lançamento das músicas')
plt.xlabel('Ano de lançamento')
plt.ylabel('Num. músicas')

plt.axvline(min_year, color = 'green', linestyle='dashed', label='min')
plt.axvline(mean_year, color = 'cyan', linestyle='dashed', label='mean')
plt.axvline(median_year, color = 'red', linestyle='dashed', label='median')
plt.axvline(max_year, color = 'orange', linestyle='dashed', label='max')

plt.legend(loc=2)
```

```python
min_pop = tiktok['popularity'].min()
max_pop = tiktok['popularity'].max()
mean_pop = tiktok['popularity'].mean()
median_pop = tiktok['popularity'].median()

print("Popularidade mínima:", min_pop)
print("Popularidade máxima:", max_pop)

tiktok.plot.hist(y='popularity', edgecolor='k')
plt.title('Histograma da popularidade das músicas')
plt.xlabel('Popularidade')
plt.ylabel('Num. músicas')

plt.axvline(min_pop, color = 'green', linestyle='dashed', label='min')
plt.axvline(mean_pop, color = 'cyan', linestyle='dashed', label='mean')
plt.axvline(median_pop, color = 'red', linestyle='dashed', label='median')
plt.axvline(max_pop, color = 'orange', linestyle='dashed', label='max')

plt.legend(loc=2)
```

```python
# Regressão linear múltipla

drop_columns = ['track_name', 'artist_name', 'release_date', 'mode', 'key', 'duration_mins', 'genre', 
               'popularity', 'liveness', 'valence', 'tempo']

X = tiktok.drop(drop_columns, axis=1)
Y = tiktok['popularity']

reg = LinearRegression()
reg.fit(X, Y)

print("O modelo é: Popularidade = {:.5} + {:.5}*duration + {:.5}*danceability + {:.5}*energy + {:.5}*loudness + {:.5}*speechiness + {:.5}*acousticness + {:.5}*instrumentalness + {:.5}*release_year".format(reg.intercept_, reg.coef_[0], reg.coef_[1], reg.coef_[2], reg.coef_[3], reg.coef_[4], reg.coef_[5], reg.coef_[6], reg.coef_[7]))
```

```python
# Avaliando a regressão múltipla

X = np.column_stack((tiktok['duration'], tiktok['danceability'], tiktok['energy'], tiktok['loudness'],
                     tiktok['speechiness'], tiktok['acousticness'], tiktok['instrumentalness'],
                     tiktok['release_year']))
y = tiktok['popularity']


X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()


print(est2.summary())
```
